```python
import pandas as pd

for col in df1.columns:
    if pd.api.types.is_numeric_dtype(df1[col]):
        # If all non‑missing values are whole numbers, cast to Int64
        if (df1[col].dropna() % 1 == 0).all():
            df1[col] = df1[col].astype("Int64")   # nullable integer
        else:
            # Otherwise keep as float (decimals preserved)
            df1[col] = df1[col].astype("float")
    else:
        # Try to coerce text columns that look numeric, leave others untouched
        df1[col] = pd.to_numeric(df1[col], errors="ignore")
```

```python
import numpy as np

# Replace the specific error string with NaN, then convert
df1['Population'] = df1['Population'].replace('#DIV/0!', np.nan).astype(float)
```
