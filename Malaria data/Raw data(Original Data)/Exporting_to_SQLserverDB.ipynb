### Step IV:  Change columns to respective data types

```python
# Changing all necessary column values to Integer 
```

```python
import pandas as pd

for col in df1.columns:
    if pd.api.types.is_numeric_dtype(df1[col]):
        # If all non‑missing values are whole numbers, cast to Int64
        if (df1[col].dropna() % 1 == 0).all():
            df1[col] = df1[col].astype("Int64")   # nullable integer
        else:
            # Otherwise keep as float (decimals preserved)
            df1[col] = df1[col].astype("float")
    else:
        # Try to coerce text columns that look numeric, leave others untouched
        df1[col] = pd.to_numeric(df1[col], errors="ignore")
```

```python
# Replace the specific error string with NaN, then convert
df1['Population'] = df1['Population'].replace('#DIV/0!', np.nan).astype(float)
```

### Step V: Push the raw files to the SQL server Database

```python
server = "DESKTOP-U507UFU\\MSSQLSERVER01"
database = "Malaria_DB"

conn_str = (
    "Driver={ODBC Driver 17 for SQL Server};"
    f"Server={server};"
    f"Database={database};"
    "Trusted_Connection=yes;"
)

params = urllib.parse.quote_plus(conn_str)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

df1.to_sql(
    "Malaria2023",
    engine,
    if_exists="append",
    index=False
)
```
