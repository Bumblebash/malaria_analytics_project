### Script to check Duplicate Columns in the CSV File

```python
import pandas as pd

# Path to the specific file giving you the SSIS error
file_path = r"D:\Data Analyst\MALARIA ANALYTICS\FULL PROJECT\Malaria data\Raw data(Original Data)\Raw data\Jamir Data 202.csv"

# Load only the header row to save time and memory
df_headers = pd.read_csv(file_path, nrows=0)

# Find duplicates
duplicate_columns = df_headers.columns[df_headers.columns.duplicated()].tolist()

# Print the results
print("--- SSIS DUPLICATE VERIFICATION REPORT ---")
if duplicate_columns:
    print(f"❌ SSIS IS RIGHT! Found {len(duplicate_columns)} exact duplicate column name(s):")
    for col in set(duplicate_columns):
        count = list(df_headers.columns).count(col)
        print(f"  • Column Name: '{col}' appears {count} times in the file.")
else:
    print("✅ SSIS IS WRONG! All column headers are structurally unique text strings.")
    print("   The issue is likely an SSIS text-truncation limit or parser bug reading the matrix layout.")
```
