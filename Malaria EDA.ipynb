```python
import pandas as pd
import os

# Create an empty list to hold dataframes
all_dfs = []

# It's good practice to print to confirm the loop is working
print("Starting to process files...")

for year in [2020, 2021, 2022, 2023, 2024]:
    # 1. Construct the file name. This looks correct based on your file explorer.
    file_path = f'Malaria data {year}.xlsx'
    print(f"Attempting to read: {file_path}")
    
    # 2. Check if the file exists before trying to read it
    if not os.path.isfile(file_path):
        print(f"Error: The file {file_path} was not found. Check the path.")
        continue # Skip to the next year if file is missing
    
    try:
        # 3. FIRST, inspect the sheet names without reading the data
        xl = pd.ExcelFile(file_path)
        print(f"Sheet names in {file_path}: {xl.sheet_names}")
        
        # 4. Assume the first sheet is the one we want. This is the most common scenario.
        #    Replace 0 with the correct sheet name from the list printed above if needed.
        sheet_name = xl.sheet_names[0] # Gets the name of the first sheet
        print(f"Using sheet: '{sheet_name}'")
        
        # 5. Now read the specific sheet
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        
        # 6. Add the year column
        df['Year'] = year
        all_dfs.append(df)
        print(f"Successfully processed {file_path}")
        
    except Exception as e:
        print(f"An error occurred with {file_path}: {e}")

# Check if we have any data before trying to combine
if all_dfs:
    # Combine all DataFrames
    master_df = pd.concat(all_dfs, ignore_index=True)
    print(f"Combined master DataFrame created with shape: {master_df.shape}")
    
    # Preview the data
    print(master_df.head())
else:
    print("No data was loaded. The master DataFrame was not created.")
```

```python
import pandas as pd
import os
import glob

# Base path
base_path = r"D:\Data Analyst\MALARIA ANALYTICS\TOTALS & CORDINATES"

# Get all Excel files in the folder
excel_files = glob.glob(os.path.join(base_path, "*.xlsx"))

all_dfs = []

for file in excel_files:
    # Ignore temporary Excel files (~$)
    if os.path.basename(file).startswith("~$"):
        continue
    
    # Extract year from filename (keep only digits)
    year = "".join(filter(str.isdigit, os.path.basename(file)))
    
    # Read the file
    df = pd.read_excel(file, sheet_name="Sheet1")
    df["Year"] = year  # Add year column
    
    all_dfs.append(df)

# Combine into one dataframe
master_df = pd.concat(all_dfs, ignore_index=True)

print("✅ Combined dataframe shape:", master_df.shape)
```
