### Merging All Spatial stat files and Adding A Year Column

```python
import glob 
import pandas as pd
import os # for path functions


def clean_columns(df):
    df.columns = df.columns.str.strip() #Removing White spaces from columns
    df.columns = df.columns.str.replace(r"\s+", " ", regex=True) #normalize multiple spaces to one
    return df 

# Collect all CSV
all_files = glob.glob("C:\\Windows\\System32\\1033\\MALARIA EDA/spatial_stats/*.csv")


dfs = []
for file in all_files:
    df = pd.read_csv(file)
    df = clean_columns(df)
    
    # Extract year from file name (this needs to be inside the loop)
    year = os.path.splitext(os.path.basename(file))[0].split("_")[-1]
    df["Year"] = int(year)
    
    dfs.append(df)  # This also needs to be inside the loop

# Combine into one master DataFrame
spatial_df = pd.concat(dfs, ignore_index=True)

print(spatial_df[["Year"]].head(20))
print(spatial_df.columns.tolist())
```

### Merging Misfit  Columns
- 10-19YRS TOTAL
- 29DAYS-4YRS MALE
- 10-19 YRS FEMALE (B/S & RDT)
- 20+ (B/S &RDT) MALE 

```python
# Dictionary of columns that should be merged
# Dictionary of columns that should be merged
merge_map = {
    "20+ (B/S &RDT) MALE": "20+ (B/S & RDT) MALE",
    "29DAYS-4YRS MALE": "29DAYS - 4YRS MALE",
    "10-19 YRS TOTAL": "10-19YRS TOTAL",
    "10-19 YRS FEMALE (B/S $ RDT )": "10-19 YRS (B/S & RDT) FEMALE"
}

for bad_col, good_col in merge_map.items():
    if bad_col in spatial_df.columns and good_col in spatial_df.columns:
        # Merge values where good_col is missing
        spatial_df[good_col] = spatial_df[good_col].fillna(spatial_df[bad_col])
        # Drop the redundant col
        spatial_df = spatial_df.drop(columns=[bad_col])
```
